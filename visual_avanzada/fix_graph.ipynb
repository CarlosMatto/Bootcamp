```python


# Cargar el dataset 
df = sns.load_dataset('tips')

# Crear una figura y un eje
fig, ax = plt.subplots(figsize=(12, 6))
palette = {'Lunch': 'blue', 'Dinner': 'red'}

# Gráfico de dispersión con cuadrícula, leyenda, anotaciones y flechas
sns.scatterplot(data=df, x='total_bill', y='tip', hue='time', style='smoker', s=100, ax=ax, palette=palette)
ax.set_title('Relación entre Total Bill y Tip con Anotaciones y Flechas', fontsize=18)
ax.set_xlabel('Total Bill ($)', fontsize=15)
ax.set_ylabel('Tip ($)', fontsize=15)
ax.legend(loc='upper right', fontsize='large', shadow=True, title='Detalles')

# Añadir cuadrícula
ax.grid(True, which='both', linestyle='--', linewidth=0.7, color='grey')

# Cambiar el fondo del gráfico a gris claro
ax.set_facecolor('#D3D3D3')  # Color gris claro

# Añadir la leyenda con transparencia
ax.legend(loc='upper right', fontsize='large', shadow=False, title='Detalles', framealpha=0)


# Añadir anotación
ax.annotate('Mayor tip en la cena', xy=(50, 10), xytext=(40, 15),
            arrowprops=dict(facecolor='black', shrink=0.05))

# Añadir flecha
ax.arrow(30, 5, 10, 3, head_width=0.8, head_length=0.5, fc='green', ec='green')

# Mostrar el gráfico
plt.show()
```
